```python
import pandas as pd
#FileName = 'C:\\00 Big Data Analytics\\03 Visual Analytics\\Assignments\\Assignment 01 Simple Visualization\\GDP Data For Countries.csv'
FileName = 'C:\\Dd\\GDPDataForCountries.csv'
GDP = pd.read_csv(FileName, skiprows=4) #, index_col = 0
df = pd.DataFrame(GDP)
df
```

```python
# Set every NaN to Zero
df.fillna(0, inplace=True)
df
```

```python
# To remove (Country Name	Indicator Name	Indicator Code).
df.drop(['Country Name', 'Indicator Name', 'Indicator Code'], axis=1, inplace=True)

# To remove the Unnamed (last) field
Pdf = df.iloc[:,0:-1]
Pdf
```

```python
# Rows = len(Pdf.axes[0])
# Rows = Pdf.shape[0]
Rows = len(Pdf.index)   # Or len(Pdf)

# Cols = len(Pdf.axes[1])
# Cols = Pdf.shape[1]
Cols = len(Pdf.columns)

print('No of Rows = ', Rows)
print('No of Cols = ', Cols)
```

```python
def AvgGDP(x, y, z):
  if ((x == 0.00) & (y == 0.00)):
    result = z
  elif ((x == 0.00) & (y != 0.00)):
    result = (y + z)/2.0
  elif ((x != 0.00) & (y == 0.00)):
    result = (x + z)/2.0
  else:
    result = (x + y + z)/3.0
  return result
```

```python
# REPLACING THE ZEROS with meaningful values.
# I choose to calculate, and use the average of the last three years' values because, I observed that the GDP for most of the
# countries is growing yearly. If I use the Pandas Mean function, it will look at the entire years for the country, which not
# give me the desired result. Rather, I call my own funtion to calculate the Mean using the most recent three years' values.

# To use the average of the last three years values as a replacement for the Zero values.
for row in range(0, Rows):
    Yr1GDP = 0
    Yr2GDP = 0
    Yr3GDP = 0
    for col in range(1, Cols):
        if (Pdf.iloc[row, col] == 0.00):
            CurGDP = AvgGDP(Yr1GDP, Yr2GDP, Yr3GDP)
            Pdf.iloc[row, col] = CurGDP
            Yr1GDP = Yr2GDP
            Yr2GDP = Yr3GDP
            Yr3GDP = CurGDP
            #print('IF Block, Col = ', col, 'Value = ', Pdf.iloc[row, col])
        else:
            Yr1GDP = Yr2GDP
            Yr2GDP = Yr3GDP
            Yr3GDP = Pdf.iloc[row, col]
            #print('ELSE Block, Col = ', col, 'Value = ', Pdf.iloc[row, col])
```

```python
# Now, with the exception of the preceeding Zeros for each of the countries, all other Zeros has been replaced
# with meaningful values.
Pdf
```
